## System Design Notes

### Selecting the best database

- generally don't affect the functional requirement 
- depends on the non-functional requirements like data structure and scalability requirements
- factors affecting the choice
    - structure / non-structured data 
    - query pattern 
    - scalability 

#### Caching solution 
- any system design solution would need some kind of caching
- you have a key and value: 
    - key: query pattern or request params
    - value: the response of the api call or query result 
- common caching solutions: `redis(battle-tested)`, `memcached`, `etcd`, `hazelcast`
- frontend caching: `Next Js`, `RTK - Query`, `Tanstack Query`

#### File Storage solution 
- When data is of type images, videos, pdfs huge documents etc
- Use blob storage systems like `Amazon S3`

#### What is a Content Delivery Network?
- The network which distributes the file in blob storage to servers across the world
- Faster than querying in S3

Blob solution example: `S3 + CDN`

#### Text Search Engine
- For more niche system design questions like E-Commerce, streaming services
- provided by `elastic search`, `solr`
- powered by `apache lucene`  
- Fuzzy Text Search: handling user misspelt words to return correct result nevertheless 
- Not recommended to use it as a primary data source

#### Time Series Database
- A kind of relational database used for metrics monitoring system 
- append-only write mode
- no random fetching of data and no random updation of data
- fetching is also time-grouped
- solutions are: `OpenTSDB`

#### Analytics on transactions 
- We need a data warehouse: a large database where we can dump all data with some query capability for analytics on transactions
- Solutions: `Hadoop`

#### Types of Databases:

- Is the Data Structured?
    - Yes
        - Do we need ACID properties?
            - Yes?
                - RDBMS: `MySQL`, `Oracle`, `Postgre SQL`
                - Payment systems, inventory management systems
                - ability for constructing complex queries and store minimum amount of data
            - No? 
                - NoSQL and SQL are fine
    - No
        - JSON type datasets for catalouging purposes: `Mongo DB`
        - In cases of non-relational datasets with absence of complex queries, we can still use `Document DB`, if it doesn't fall i next category
        - In case of ever increasing data and finite number of queries based on the almost-infinite data: `Columnar DB` like `Cassandra` and `HBase`.
    - If we don't need ACID properties, Relational datasets, or ever-increasing databases, usage of any of these is fine

### Complex Database Example: 

- Amazon has ever increasing number of orders and with the need for ACID properties, so for newer transactions, `Relational DB` is used and later on put on permanant store in `Columnar DB`.
